```python
import random as rand
com = rand.randint(1,3)
user = int(input())

# 1 : 가위 / 2 : 바위 / 3 : 보
d = {"1" : "가위", "2" : "바위", "3": "보"}
print("COM(%s) : USER(%s)" %(d[str(com)], d[str(user)]))

if com == user:
  print("DRAW")
elif(com == 1 and user == 3) or (com == 2 and user == 1) or (com == 3 and user == 2):
  print("COM WINS")
else:
  print("USER WINS")
```

```python
import random as rand
num1 = rand.randint(1,6)
num2 = rand.randint(1,6)
num3 = rand.randint(1,6)
num4 = rand.randint(1,6)

print(num1, num2, num3, num4)
user1_success = False
user2_success = False

if num1 + num2 >= 10:
  user1_success = True

if num3 + num4 >= 10:
  user2_success = True

if user1_success and user2_success:
  if (num1 + num2) > (num3 + num4):
    print("User 1 wins")
  elif (num1 + num2) < (num3 + num4):
    print("User 2 wins")
  else:
    print("Draw")
elif user1_success and not user2_success:
  print("User 1 wins")
else:
  print("User 2 wins")
```

```python
from random import choice, randint

fruit = ["apple", "banana", "strawberry", "watermelon", "mandarin",
         "peach", "grapes", "orange", "pear", "kiwi"]

score = 0
n = randint(5, 10)                       # 문제 개수 (5~10)

for i in range(n):
    s = choice(fruit)                    # 단어 랜덤 선택
    word = input('[%d/%d] 다음 단어를 입력하세요 -> %s\n' % (i + 1, n, s))
    if s == word:                        # 정답이면 점수 증가
        score += 1

rate = score / n * 100                   # 정확도 계산
print('%d/%d번 성공!' % (score, n))
print('정확도 : %.1f%%' % rate)

if rate == 100:
    print('집에 데려다줘서 고마워!!')
elif rate >= 80:
    print('집이 코앞인데!! 한 번만 더 시도해줘!!')
elif rate >= 50:
    print('집에 가고 싶어!!')
else:
    print('거북이가 쓰러졌어요...')
```
