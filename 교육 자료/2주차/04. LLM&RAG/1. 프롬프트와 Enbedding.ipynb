```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

load_dotenv()

llm = ChatOpenAI(
    model='gpt-5-mini',
    temperature=0
)
```

- temperature는 값이 낮을수록 LLM이 사실적인 답변만 제시함
  - 정확한 답, 코드 => 0 ~ 0.2
  - RAG와 같은 다른 저오를 참고해서 답변을 할때 => 0.2 ~0.5
  - 카피라이트 => 0.7 ~ 1.0
  - 1.0 이상은 거의 사용하지 않음

### LLM 입력을 위한 프롬프트(Prompt) 생성
1. System Prompt(시스템 메세지) => LLM에게 규칙/규정/결과 형식 등을 동작을 지시하는 내용
2. User Prompt(사용자 메세지) => LLM에게 전달할 데이터
3. AI Prompt(AI 메세지) => LLM이 이전 답변을 다시 LLM에게 전달

```python
# 기본 프롬프트
message="RAG에 대해 간단히 설명해주세요"
ai_msg = llm.invoke(message)
ai_msg.content
```

## 코사인 유사도
- 측정 방식: 두 벡터 사이의 코사인 각도를 계산.
- 결과 범위: 값은 -1에서 1 사이의 값을 가진다
  - 1에 가까울수록 두 벡터는 같은 방향(매우 유사)
  - 0에 가까울수록 직교(연관성 없음)
  - 1에 가까울수록 반대 방향(정반대)


```python
# 유사도 검색 -> 코사이 유사도

import numpy as np
from sklearn.metrics.pairwise  import cosine_similarity

v1 = np.array([[1, 2, 3]])
v2 = np.array([[10, 20, 30], [-5, -6, -7]])

cos_sim = cosine_similarity(v1, v2)
print(cos_sim)
```

### Embedding Vector 만들기
1. 무료를 이용한 모델 사용: 저차원 벡터 생성 => 표현에 한계가 있음
2. Open API Embedding Model(유료): 고차원 벡터 생상 => 다양한 표현이 가능함

### 무료: SentenceTransformer
- 텍스트(문장, 단락, 문단 등)를 고정된 크기의 밀집 벡터(Dense Vector)로 변환해 주는 파이썬 라이브러리
- 단순히 단어의 빈도를 세는 방식(TF-IDF 등)과 달리, 문장의 의미적 맥락을 파악하여 벡터로 표현하는 것이 특징
- 주요 특징
  - 의미적 유사도(Semantic Similarity): "점심 메뉴로 무엇이 좋을까?"와 "오늘 점심은 뭘 먹지?"라는 두 문장은 단어 구성이 완전히 다르지만, 의미는 같습니다. SentenceTransformer는 이 두 문장을 비슷한 위치의 벡터로 변환하여, 코사인 유사도 계산 시 높은 유사도를 보여줍니다.
  - 임베딩 벡터 생성: 텍스트를 고차원 숫자 배열로 변환함으로써, 벡터 데이터베이스(Vector DB)에 저장하여 빠른 검색이 가능하게 합니다. 이는 최근의 RAG(Retrieval-Augmented Generation) 시스템 구축의 핵심 요소입니다.
  - 간편한 사용: Hugging Face에 공개된 수많은 모델을 단 몇 줄의 코드로 불러와서 즉시 사용할 수 있습니다.
- 사용 이유
  - 고차원 표현: 기존의 단순 벡터화 방식보다 훨씬 풍부한 의미 정보(고차원)를 담고 있습니다.
  - 다양한 언어 지원: 한국어 데이터로 특화 학습된 모델(ko-sroberta 등)을 사용하면 한국어의 미묘한 의미 차이까지 잘 잡아낼 수 있습니다.
