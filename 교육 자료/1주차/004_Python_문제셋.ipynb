**문제 5-3 [고급]**

정수 n을 입력받아 소수(1과 자기 자신만으로 나누어짐)면 `소수`, 아니면 `합성수` 를 출력하세요. (break로 조기 종료)

````python
# 여기에 코드를 작성하세요
n = int(input())

if n % 2 == 0:
  print("소수 아님")
else:
  is_prime = True
  i = 2
  while i < n:
    if n % i == 0:
      is_prime = False
      break
    i += 1
  if is_prime:
    print("소수")
  else:
    print("합성수")
````

```python
            is_prime = False
            break

    else:
        print("합성수")
```

## 11. 종합 문제

**문제 11-1 [중급]**

리스트 `[80, 95, 72, 88, 90]` 의 최고점, 최저점, 평균을 반복문으로 구해 출력하세요. (max/min/sum 없이)

````python
# 여기에 코드를 작성하세요
````

<details>
<summary><b>정답 예시 보기</b></summary>

```python
scores = [80, 95, 72, 88, 90]
high = scores[0]
low = scores[0]
total = 0
for s in scores:
    if s > high:
        high = s
    if s < low:
        low = s
    total += s
print("최고:", high)
print("최저:", low)
print("평균:", total / len(scores))
```

</details>

**문제 11-2 [고급]**

학생별 점수 리스트 딕셔너리 `{"철수": [90, 80], "영희": [100, 90]}` 에서 학생별 평균을 출력하세요.

```python
students = {"철수": [90, 80], "영희": [100, 90]}

    nums = students[name]
```

**문제 11-3 [고급]**

문자열 `"apple banana apple cherry banana apple"` 에서 가장 많이 나온 단어와 횟수를 출력하세요. (결과: `apple 3`)

````python
# 여기에 코드를 작성하세요
````

<details>
<summary><b>정답 예시 보기</b></summary>

```python
text = "apple banana apple cherry banana apple"
words = text.split()
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1
best = ""
cnt = 0
for w in freq:
    if freq[w] > cnt:
        cnt = freq[w]
        best = w
print(best, cnt)
```
