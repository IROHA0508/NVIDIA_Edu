# 랭체인으로 연결

- HumanMessage : 사람이 입력값으로 넣은 질문
- AIMessage : 모델 호출의 결과 -> 즉 모델이 생성한 출력
- SystemMessage : 시스템 프롬프트 / 모델의 태도, 역할, 규칙을 생성
- ToolMessage : tool의 호출 결과를 모델에 다시 전달할 떄 사용

```python
from langchain_openai import ChatOpenAI
from langchain.messages import HumanMessage, AIMessage, SystemMessage

model = ChatOpenAI(model="gpt-5.6")

model.invoke([HumanMessage("안녕! 랭체인에 대해 설명해줘")])
```

```python
# Langchain Docs에서 설명하는 init_chat_model로 모델 생성
import os
from langchain.chat_models import init_chat_model

model = init_chat_model(
        "gpt-5.6",
        temperature = 0.4
        )

response = model.invoke("만화 블리치에 대해서 알아?")
response
```

- LangChain에서 Prompt 추가

```python
messages = [
  # SystemMessage를 이용해서 Role 부여
  SystemMessage("너는 애니메이션/만화 전문가야. 스포일러 없이 핵심만 간단히 설명해줘."),
  HumanMessage("만화 블리치 주인공인 이치고에 대해서 설명해줘")
]

response = model.invoke(messages)
response.content
```

- ChatPromptTemplate을 이용해서 Prompt를 다양하게 만들기

```python
from langchain_core.prompts import ChatPromptTemplate

# ChatPromptTemplate을 이용하면 메세지 내부에 변수 자리를 마련해놓을 수 있다
prompt = ChatPromptTemplate([
    ("system", "너는 {genre} 전문가야. 스포일러 없이 핵심만 간단히 설명해줘."),
    ("system", "답변을 출력할 때 Markdown 양식 말고 순수 텍스트로만 출력할 것"),
    ("human", "{title}에 대해서 알아?"),
])

# 프롬프트를 실제 메시지로 변환
prompt_value = prompt.invoke({"genre": "만화", "title": "원피스"})

response = model.invoke(prompt_value)
response.content
```

- Chaining으로 모델과 프롬프트 연결하기   
prompt | model | output_parser 방식으로 데이터를 왼쪽에서 오른쪽으로 보내는 느낌

```python
# 프롬프트와 모델을 분리해서 구현
chained_model = prompt | model
response = chained_model.invoke({"genre" : "만화", "title" : "원피스"})
response.content
```

- output_parser 까지 연결해보기

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

full_model = prompt | model | parser

# parser를 연결해서 response.content를 안해도 텍스트만 불러온다
full_model.invoke({"genre" : "만화", "title" : "원피스"})
```

# 코사인 유사도

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

v1 = np.array([
  [1, 2, 3],
  [4, 5, 6]
])

v2 = np.array([
  [10, 20, 30],
  [-20, -50, -70]
])

cos_sim = cosine_similarity(v1, v2)
print(cos_sim)
```

```python
# 무료 버전을 이용한 벡터화
from sentence_transformers import SentenceTransformer

# 1. 모델 불러오기
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

# 2. 임베딩 생성할 문장들 => 문장 사전
sentences = [
    "고양이가 창가에서 햇볕을 쬐고 있다",
    "새끼 고양이가 실타래를 가지고 논다",
    "강아지가 공원에서 뛰어다닌다",
    "강아지가 주인을 보고 꼬리를 흔든다",
    "호랑이가 숲속을 걸어 다닌다",
    "사자가 초원에서 사냥을 준비한다",
    "코끼리가 긴 코로 나뭇가지를 잡는다",
    "기린이 나무 위 잎을 먹는다",
    "원숭이가 바나나를 먹는다",
    "펭귄이 얼음 위에서 걸어간다",
    "사과가 빨갛게 익었다",
    "바나나가 노랗게 잘 익어 달콤하다",
    "포도가 달콤하고 신선하다",
    "빵이 오븐에서 갓 구워졌다",
    "라면이 뜨겁게 끓고 있다",
    "김치찌개에서 얼큰한 냄새가 난다",
    "초밥이 접시에 예쁘게 놓여 있다",
    "피자가 치즈로 가득 덮여 있다",
    "햄버거에 고기 패티가 두껍다",
    "커피 향이 방 안에 가득 퍼졌다",
    "서울의 남산타워가 빛나고 있다",
    "한강 다리 위에서 야경을 본다",
    "제주도의 바닷가가 파랗게 빛난다",
    "부산 해운대 해수욕장에 사람이 많다",
    "경복궁의 전통 건물이 웅장하다",
    "뉴욕의 자유의 여신상이 우뚝 서 있다",
    "파리의 에펠탑이 밤에 반짝인다",
    "런던의 빅벤 시계탑이 울린다",
    "도쿄의 시부야 거리가 붐빈다",
    "로마의 콜로세움이 고대의 흔적을 보여준다"
]
```

```python
embedding = model.encode(sentences)
print(f'임베딩 벡터 수 : {len(embedding)}, 차원 : {embedding.shape}')
```

### embedding vector 결과(위치) 확인하기
- TensorFlow Embedding Projector(https://projector.tensorflow.org/): 구글에서 제공하는 웹 사이트를 통해 직접 시각화
  - 1단계: Colab에서 생성한 embeddings 배열을 tsv 파일로 저장
  - 2단계: Embedding Projector 접속 후, 좌측 상단의 [Load] 버튼을 클릭
  - 3단계: Step 1에서 생성한 vectors.tsv를 Step 1(Load a TSV file of vectors)에, metadata.tsv를 Step 2(Load a TSV file of metadata)에 각각 업로드
  - 4단계: 우측의 'T-SNE'나 'PCA'를 선택하여 데이터가 어떻게 뭉쳐 있는지 확인


```python
# 4. vectors.tsv 저장

with open('../실습 자료/vectors.tsv', 'w', encoding='utf-8') as f:
    for emb in embedding:
        f.write('\t'.join([str(x) for x in emb])+ '\n')

# 5. metadata.tsv 저장
with open('../실습 자료/metadata.tsv', 'w', encoding='utf-8') as f:
    for s in sentences:
        f.write(s + '\n')
```

```python
# 6. 결과 확인
query = input('문장을 입력하세요.')

query_embedding = model.encode([query])

similarity = cosine_similarity(query_embedding, embedding)[0]
similarity
```

```python
# 유사도가 높은 상위 2개의 문장 찾기
top2_idx = similarity.argsort()[-2:][::-1]
print(top2_idx)
```

```python
print(f"입력된 문장 : {query}")

for idx in top2_idx:
  print(f"문장 : {sentences[idx]} /// 유사도 : {similarity[idx]:.4f}")
```
