# ліҙмҠӨн„ҙ мЈјнғқ к°ҖкІ© мҳҲмёЎ

```python
from tensorflow.keras.datasets.boston_housing import load_data
import numpy as np

# лҚ°мқҙн„°лҘј лӢӨмҡҙл°ӣмҠөлӢҲлӢӨ.
(X_train, Y_train), (X_test, Y_test) = load_data(path='boston_housing.npz',
                                                 test_split=0.2,
                                                 seed=777)

print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

# лҚ°мқҙн„° м „мІҳлҰ¬

```python
import numpy as np

# лҚ°мқҙн„° н‘ңмӨҖнҷ”
# мӮ¬мқҙнӮ·лҹ°м—җм„ң standard scaler мӮ¬мҡ©н•ҙлҸ„ лҗЁ
mean = np.mean(X_train, axis = 0)
std = np.std(X_train, axis = 0)

x_train = (X_train - mean) / std
x_test = (X_test - mean) / std

# нӣҲл Ё лҚ°мқҙн„°м…Ӣкіј кІҖмҰқ лҚ°мқҙн„°м…ӢмңјлЎң лӮҳлҲ•лӢҲлӢӨ.
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, Y_train,
                                                  test_size = 0.33,
                                                  random_state = 777)

# мӮ¬мқҙнӮ·лҹ°мқ„ мқҙмҡ©н•ң н‘ңмӨҖнҷ” (X_train, X_testлҠ” мң„м—җ м—Ҷм§Җл§Ң кө¬л¶„н•ҳкё° мң„н•ҙ мӮ¬мҡ©н•Ё)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler_model = scaler.fit(X_train)
X_train_scaled = scaler_model.transform(X_train)
X_test_scaled = scaler_model.transform(X_test)
```

# лӘЁлҚё кө¬м„ұ
- лӘЁлҚёмқҳ л§Ҳм§Җл§ү Dense мёөм—җм„ң лі„лҸ„мқҳ нҷңм„ұнҷ” н•ЁмҲҳлҘј мӮ¬мҡ©н•ҳм§Җ м•ҠмқҢ
  - мқёмһҗлҘј м„Өм •н•ҳм§Җ м•ҠмқҖ кІҪмҡ°, defaultлҠ” вҖҳlinearвҖҷлЎң м„Өм •
- мҶҗмӢӨ н•ЁмҲҳлҠ” нҡҢк·Җ л¬ём ңм—җм„ң мЈјлЎң мӮ¬мҡ©лҗҳлҠ” нҸүк·  м ңкіұ мҳӨм°Ё(MSE; Mean Squared Error)лҘј мӮ¬мҡ©


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

# нҡҢк·ҖлҘј л”Ҙлҹ¬лӢқмңјлЎң кө¬нҳ„н•  кІҪмҡ° л§Ё л§Ҳм§Җл§үм—җ activation functionмқҙ мЎҙмһ¬н•ҳм§Җ м•ҠлҠ”лӢӨ O_O
# мһ…л Ҙ лҚ°мқҙн„°мқҳ нҳ•нғңлҘј кјӯ лӘ…мӢңн•ҙм•ј н•©лӢҲлӢӨ.
# 13м°Ёмӣҗмқҳ лҚ°мқҙн„°лҘј мһ…л ҘмңјлЎң л°ӣкі , 64к°ңмқҳ м¶ңл Ҙмқ„ к°Җм§ҖлҠ” мІ« лІҲм§ё Dense мёө
model.add(Dense(64, activation = 'relu', input_shape = (13, )))
model.add(Dense(32, activation = 'relu')) # 32к°ңмқҳ м¶ңл Ҙмқ„ к°Җм§ҖлҠ” Dense мёө
model.add(Dense(1)) # н•ҳлӮҳмқҳ к°’мқ„ м¶ңл Ҙн•©лӢҲлӢӨ. -> мөңмў… к°ҖкІ©мқ„ м¶ңл Ҙн•ҙм•јн•Ё

model.compile(optimizer = 'adam', loss = 'mse', metrics = ['mae'])
```

```python
history = model.fit(x_train, y_train,
                    epochs = 300,
                    validation_data = (x_val, y_val))
```

# K-Fold мӮ¬мҡ©н•ҳкё°

# DataSet лӢӨмӢң лЎңл“ң, н‘ңмӨҖнҷ”

```python
from tensorflow.keras.datasets.boston_housing import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

(x_train, y_train), (x_test, y_test) = load_data(path='boston_housing.npz',
                                                 test_split=0.2,
                                                 seed=777)

# лҚ°мқҙн„° н‘ңмӨҖнҷ”
mean = np.mean(x_train, axis = 0)
std = np.std(x_train, axis = 0)
# м—¬кё°к№Ңм§„ м „л¶Җ лҸҷмқјн•©лӢҲлӢӨ.
x_train = (x_train - mean) / std
x_test = (x_test - mean) / std
```

# K-Fold м •мқҳ

```python
from sklearn.model_selection import KFold
#----------------------------------------
# K-FoldлҘј м§„н–үн•ҙлҙ…лӢҲлӢӨ.
k = 3

# мЈјм–ҙм§„ лҚ°мқҙн„°м…Ӣмқ„ kл§ҢнҒј л“ұл¶„н•©лӢҲлӢӨ.
# м—¬кё°м„ңлҠ” 3мқҙлҜҖлЎң нӣҲл Ё лҚ°мқҙн„°м…Ӣ(404к°ң)лҘј 3л“ұл¶„н•ҳм—¬
# 1к°ңлҠ” кІҖмҰқм…ӢмңјлЎң, лӮҳлЁём§Җ 2к°ңлҠ” нӣҲл Ём…ӢмңјлЎң нҷңмҡ©н•©лӢҲлӢӨ.
kfold = KFold(n_splits=k, random_state = 777, shuffle = True)

# мһ¬мӮ¬мҡ©мқ„ мң„н•ҙ лӘЁлҚёмқ„ л°ҳнҷҳн•ҳлҠ” н•ЁмҲҳлҘј м •мқҳн•©лӢҲлӢӨ.
# мң„м—җм„ң мӮ¬мҡ©н–ҲлҚҳ лӘЁлҚёкіј лҸҷмқјн•Ё
def get_model():
    model = Sequential()
    model.add(Dense(64, activation = 'relu', input_shape = (13, )))
    model.add(Dense(32, activation = 'relu'))
    model.add(Dense(1))

    model.compile(optimizer = 'adam', loss = 'mse', metrics = ['mae'])

    return model

mae_list = [] # н…ҢмҠӨнҠём…Ӣмқ„ нҸүк°Җн•ң нӣ„ кІ°кіј maeлҘј лӢҙмқ„ лҰ¬мҠӨнҠёлҘј м„ м–ён•©лӢҲлӢӨ.

# kлІҲ м§„н–үн•©лӢҲлӢӨ.
for train_index, val_index in kfold.split(x_train):
    # н•ҙлӢ№ мқёлҚұмҠӨлҠ” л¬ҙмһ‘мң„лЎң мғқм„ұлҗ©лӢҲлӢӨ.
    # л¬ҙмһ‘мң„лЎң мғқм„ұн•ҙмЈјлҠ” кІғмқҖ кіјлҢҖм Ғн•©мқ„ н”јн•  мҲҳ мһҲлҠ” мўӢмқҖ л°©лІ•мһ…лӢҲлӢӨ.
    x_train_fold, x_val_fold = x_train[train_index], x_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # лӘЁлҚёмқ„ л¶Ҳлҹ¬мҳөлӢҲлӢӨ.
    model = get_model()

    model.fit(x_train_fold, y_train_fold, epochs = 300, validation_data = (x_val_fold, y_val_fold))

    test_mse, test_mae = model.evaluate(x_test, y_test)
    mae_list.append(test_mae)
```

```python
print(f'м „мІҙ кІ°кіј: {mae_list}')
print(f'нҸүк· лӮё кІ°кіјлҘј мөңмў… кІ°кіјлЎң мӮ¬мҡ©н•©лӢҲлӢӨ: {np.mean(mae_list)}')
```
