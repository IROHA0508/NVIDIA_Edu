```python
import re
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
train_url = "https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt"
test_url  = "https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt"

train_df = pd.read_table(train_url).dropna(subset=["document"])
test_df  = pd.read_table(test_url).dropna(subset=["document"])

display(train_df.head())
display(test_df.head())
```

```python
X_train = train_df["document"].astype(str).tolist()
y_train = train_df["label"].values

X_test = test_df["document"].astype(str).tolist()
y_test = test_df["label"].values
```

```python
def clean_text(text):
  txt = re.sub(r"[^к°Җ-нһЈa-zA-Z]", " ", text)
  txt = re.sub(r"\s+", " ", txt).strip()
  return txt

train_txt = [clean_text(text) for text in X_train]
test_txt  = [clean_text(text) for text in X_test]
```

```python
vocab_size = 20000

tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(train_txt)

x_train = tokenizer.texts_to_sequences(train_txt)
x_test  = tokenizer.texts_to_sequences(test_txt)

max_len = 100

x_train = pad_sequences(x_train, maxlen=max_len, padding='post')
x_test  = pad_sequences(x_test, maxlen=max_len, padding='post')

x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train,
                                            test_size=0.2,
                                            random_state=42)

x_tr.shape, x_val.shape
```

# лӘЁлҚё мғқм„ұ 
- мң—л¶Җл¶„мқҖ 260722мӢӨмҠө(3)кіј лҸҷмқј / м•„лһҳ лӘЁлҚё кө¬нҳ„м—җм„ң лӢ¬лқјм§җ

```python
tf.random.set_seed(42)

# кё°мЎҙ SimpleRNN(64)м—җм„ң LSTM(64)лЎң ліҖкІҪ
model = Sequential([
    Embedding(input_dim=vocab_size,
              output_dim=128,
              input_length=max_len),
    LSTM(64),          # RNN н•өмӢ¬ л Ҳмқҙм–ҙ
    Dropout(0.3),
    Dense(1, activation="sigmoid")  # мқҙм§„л¶„лҘҳ
])

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])
model.summary()
```

```python
# к°•м ң мў…лЈҢ мҳөм…ҳ

# 10лІҲ л°ҳліөн•ҳлҠ” лҸҷм•Ҳ val_loss лӘЁлӢҲн„° -> к°ҖмһҘ мўӢмқҖ к°ҖмӨ‘м№ҳ ліөмӣҗ
es = EarlyStopping(monitor = 'val_loss', patience = 10, restore_best_weights = True)

# мөңм Ғмқҳ к°ҖмӨ‘м№ҳ мһҗлҸҷ м ҖмһҘ
cp = ModelCheckpoint('./models/0722_best_model.keras', monitor = 'val_loss', save_best_only = True)
```

```python
history = model.fit(
  x_tr, y_tr,
  validation_data = (x_val, y_val),
  epochs = 30,
  batch_size = 255,
  callbacks = [es,cp],
  verbose = 1
)
```

```python
import matplotlib.pyplot as plt

his_dict = history.history
loss = his_dict['loss']
val_loss = his_dict['val_loss']

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize = (10, 5))

# нӣҲл Ё л°Ҹ кІҖмҰқ мҶҗмӢӨ к·ёлҰ¬кё°
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color = 'blue', label = 'train_loss')
ax1.plot(epochs, val_loss, color = 'orange', label = 'val_loss')
ax1.set_title('train and val loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

acc = his_dict['acc']
val_acc = his_dict['val_acc']

# нӣҲл Ё л°Ҹ кІҖмҰқ м •нҷ•лҸ„ к·ёлҰ¬кё°
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, acc, color = 'blue', label = 'train_acc')
ax2.plot(epochs, val_acc, color = 'orange', label = 'val_acc')
ax2.set_title('train and val acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('acc')
ax2.legend()

plt.show()
```

```python
# Test м…ӢмңјлЎң лӘЁлҚё кІҖмҰқ
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"NSMC Test Loss: {loss:.4f}, Test Acc: {acc:.4f}")
```

```python
def predict_sentiment_kor(text: str):
    text = clean_text(text)
    seq = tokenizer.texts_to_sequences([text])
    seq = pad_sequences(seq, maxlen=max_len, padding="post")
    prob = float(model.predict(seq, verbose=0)[0][0])
    label = "кёҚм •" if prob >= 0.5 else "л¶Җм •"
    print(f"[{label}] prob={prob:.3f} | {text}")

predict_sentiment_kor("м—°кё°к°Җ л„Ҳл¬ҙ мўӢкі  мҠӨнҶ лҰ¬лҸ„ к°җлҸҷм Ғмқҙм—Ҳм–ҙмҡ”!!!!!!!")
predict_sentiment_kor("м§ҖлЈЁн•ҳкі  мӢңк°„ м•„к№Ңмҡҙ мҳҒнҷ”мҳҖмҠөлӢҲлӢӨ")
```
