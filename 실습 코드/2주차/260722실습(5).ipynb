```python
import os
from dotenv import load_dotenv
import requests

load_dotenv()
api_key = os.getenv('WEATHER_API_KEY')


start_date = 20260721
end_date = 20260721
start_hour = 9
end_hour = 18
query_date = f'endDt={end_date}&endHh={end_hour}&startHh={start_hour}&startDt={start_date}'
print(query_date)

url = f'https://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList?serviceKey={api_key}&numOfRows=10&pageNo=1&dataType=JSON&dataCd=ASOS&dateCd=HR&stnIds=108&'
url += query_date

res = requests.get(url)

if res.status_code == 200:
  data = res.json()
  print(data)
```

```python
json_data = data['response']['body']['items']['item']
json_data[0]
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(json_data[0], orient = 'index', columns = ['데이터'])
df = df.replace('','Null')
df
```
