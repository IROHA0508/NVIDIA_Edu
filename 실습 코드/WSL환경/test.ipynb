```python
# 파이토치 테스트
import torch

print(torch.__version__)
print(torch.cuda.is_available())
print(torch.backends.cudnn.enabled)  # True 나와야 함
print(torch.backends.cudnn.version())  # 8902 같은 버전 숫자
print(torch.cuda.get_device_name(0))
```

```python
# 텐서플로우 테스트
import tensorflow as tf

print("TensorFlow 버전:", tf.__version__)
print("GPU 인식 여부:", tf.config.list_physical_devices('GPU'))
print("CUDA 빌드 여부:", tf.test.is_built_with_cuda())
print("사용 가능한 디바이스 목록:")
for device in tf.config.list_physical_devices():
    print(" -", device)
```
